### Notebook de validação “PyTab Technical Validation.ipynb”

Este notebook executa todos os cálculos oficiais em Python puro (sem PyTab)
e compara automaticamente com o PyTab.

```python
# ================================================================
# PyTab - Notebook de Validação Técnica
# ================================================================

import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

print("Notebook de validação técnica PyTab iniciado.")

# ===============================
# Funções auxiliares gerais
# ===============================

def validar_estatisticas(df, col):
    print("\n=== Estatísticas Descritivas ===")
    print("Média (pandas):         ", df[col].mean())
    print("Mediana:                ", df[col].median())
    print("Desvio padrão (sample):", df[col].std())
    print("Coef. de variação:      ", df[col].std() / df[col].mean())

def validar_outliers_zscore(df, col, threshold=3):
    z = np.abs(stats.zscore(df[col], nan_policy='omit'))
    out = df[z > threshold]
    print("\n=== Outliers (Z-score) ===")
    print(f"Qtd outliers > {threshold}: {len(out)}")
    return out

def validar_regressao(df, x, y):
    model = smf.ols(f"{y} ~ {x}", data=df).fit()
    print("\n=== Regressão Linear ===")
    print(model.summary())
    return model

def validar_anova(df, num, cat):
    modelo = smf.ols(f"{num} ~ C({cat})", data=df).fit()
    tabela = sm.stats.anova_lm(modelo, typ=2)
    print("\n=== ANOVA ===")
    print(tabela)
    return tabela

def validar_teste_t_2_amostras(df, col, grupo):
    g = df[grupo].unique()
    a = df[df[grupo] == g[0]][col]
    b = df[df[grupo] == g[1]][col]
    print("\n=== Teste t - 2 amostras ===")
    print(stats.ttest_ind(a, b, equal_var=False))

print("Funções carregadas.")
```
